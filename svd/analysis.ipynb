```python
import plotly.express as px
import plotly.graph_objects as go

# First trace from df_4
import plotly.graph_objects as go

fig = go.Figure()


fig.update_layout(
    title="Accuracy v/s Layer (100 generations on TriviaQA with layerwise ablation)",
    showlegend=True
)

# fig.show()

# rename the single trace for the legend
# Add a horizontal baseline
fig.add_hline(
    y=0.50,
    line_dash="dash",
    line_color="red",
    annotation_text="Baseline Accuracy",
    annotation_position="top right"
)

# Add additional dataframes as new traces
fig.add_trace(go.Scatter(
    x=df_random_baseline["layer"],
    y=df_random_baseline["accuracy"],
    mode="lines",
    name="random baseline"
))
fig.add_trace(go.Scatter(
    x=df_5["layer"],
    y=df_5["accuracy"],
    mode="lines",
    name="5 singular vectors"
))

fig.add_trace(go.Scatter(
    x=df_10["layer"],
    y=df_10["accuracy"],
    mode="lines",
    name="10 singular vectors"
))

fig.add_trace(go.Scatter(
    x=df_15["layer"],
    y=df_15["accuracy"],
    mode="lines",
    name="15 singular vectors"
))


fig.add_trace(go.Scatter(
    x=df_20["layer"],
    y=df_20["accuracy"],
    mode="lines",
    name="20 singular vectors"
))


fig.add_trace(go.Scatter(
    x=df_50["layer"],
    y=df_50["accuracy"],
    mode="lines",
    name="50 singular vectors"
))


fig.update_yaxes(range=[0, 1])

fig.show()
```
