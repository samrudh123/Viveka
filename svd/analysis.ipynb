```python
import re
import ast
from typing import Union, List, Any
from rapidfuzz import fuzz   # or from fuzzywuzzy import fuzz

def check_correctness(
    model_answer: str,
    correct_answers: Union[str, List[Any]],
    match_type: str = "word_boundary",  # now also supports "fuzzy"
    fuzzy_threshold: int = 90           # similarity % required for fuzzy
) -> int:
    """
    Returns 1 if model_answer matches any correct answer according to match_type,
    else 0. Supports: exact, word_boundary, substring, numeric, fuzzy.
    """

    if not isinstance(model_answer, str):
        return 0

    # Parse correct_answers if it's a string representation of list
    if isinstance(correct_answers, str):
        try:
            parsed = ast.literal_eval(correct_answers)
            correct_answers = parsed if isinstance(parsed, list) else [correct_answers]
        except (ValueError, SyntaxError):
            correct_answers = [correct_answers]

    if not isinstance(correct_answers, list):
        correct_answers = [correct_answers]

    model_answer_clean = model_answer.strip().lower()

    for answer in correct_answers:
        answer_str = str(answer).strip().lower()

        if match_type == "exact":
            if answer_str == model_answer_clean:
                return 1

        elif match_type == "word_boundary":
            pattern = r'\b' + re.escape(answer_str) + r'\b'
            if re.search(pattern, model_answer_clean):
                return 1

        elif match_type == "substring":
            if answer_str in model_answer_clean:
                return 1

        elif match_type == "numeric":
            model_numbers = re.findall(r'-?\d+\.?\d*', model_answer_clean)
            answer_numbers = re.findall(r'-?\d+\.?\d*', answer_str)
            if model_numbers and answer_numbers:
                try:
                    if float(model_numbers[0]) == float(answer_numbers[0]):
                        return 1
                except ValueError:
                    pass

        elif match_type == "fuzzy":
            # Partial ratio catches substring-ish matches, ratio is stricter
            score = fuzz.partial_ratio(answer_str, model_answer_clean)
            if score >= fuzzy_threshold:
                return 1

    return 0
```

```python
import json
import pandas as pd
json_path = "layer_ablation_detailed_results_5_SV.json"          # JSON with keys "0"..."25"
csv_path  = "C:/Users/arshm/Documents/Code/Viveka/linear_experiment_2_NN_Probing/datasets/triviaqa-subsampled_20k.csv"    # CSV with a 'correct_answer' column

# Load files
with open(json_path, "r", encoding="utf-8") as f:
    all_layers = json.load(f)
df_correct = pd.read_csv(csv_path)

layer_rows = []

# Iterate over every layer key (e.g. "0", "1", ... "25")
for layer_key, questions in all_layers.items():
    correct_count = 0
    total = 0

    for item in questions:
        qid = int(item["question_idx"])          # row index in CSV
        model_ans = item["generated_answer"]

        if 0 <= qid < len(df_correct):
            correct_ans = df_correct.loc[qid, "correct_answer"]
            correct_count += check_correctness(model_ans, correct_ans,
                                              match_type="word_boundary")
            total += 1

    accuracy = correct_count / total if total else 0
    layer_rows.append({"layer": layer_key, "accuracy": accuracy})

# Save final CSV with two columns: layer and accuracy
out_df = pd.DataFrame(layer_rows)
out_df.to_csv("layer_accuracy.csv", index=False)
print("Saved to layer_accuracy.csv")
```
