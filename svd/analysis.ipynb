```python
import numpy as np
import plotly.express as px
import pandas as pd
```

```python
df = pd.read_csv('layer_ablation_results_triviaqa_10sv.csv')
df.head()
```

```python
fig = px.line(df, x="layer", y="accuracy", title="Accuracy v/s Layer with 100 generations (greedy) ablating 10 SVs")
fig.add_hline(y=0.52, line_dash="dash", line_color="red",
              annotation_text="Baseline Accuracy", annotation_position="bottom right")
fig.show()
```
