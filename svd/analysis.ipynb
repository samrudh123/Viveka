```python
import json
import pandas as pd
import plotly.express as px
import textstat

# --- Load JSON ---
with open("layer_ablation_detailed_results_random_baseline.json", "r") as f:
    raw = json.load(f)

rows = []

# --- Iterate over layers ---
for layer_key, entries in raw.items():
    layer_num = int(layer_key)      # keys are "0", "1", ...
    scores = []
    for item in entries:
        text = item.get("generated_answer", "")
        # Compute Flesch Reading Ease (higher = easier to read)
        scores.append(textstat.flesch_reading_ease(text))
    if scores:  # avoid empty layer
        rows.append({"layer": layer_num, "avg_score": sum(scores) / len(scores)})

df = pd.DataFrame(rows).sort_values("layer")

# --- Plot ---
fig = px.line(
    df,
    x="layer",
    y="avg_score",
    markers=True,
    title="Average Flesch Reading Ease vs Layer"
)
fig.update_yaxes(title="Average Reading Ease (0–100 scale)")
fig.update_xaxes(title="Layer", dtick=1)
fig.update_yaxes(range=[0, 100])
fig.show()
```

```python
import torch
import re
from sentence_transformers import SentenceTransformer, util
model = SentenceTransformer("all-mpnet-base-v2")

def coherence_score(sentence):
    parts = [c.strip() for c in re.split(r'[.;:!?]', sentence) if c.strip()]
    if len(parts) < 2: return 1.0
    embeds = model.encode(parts, convert_to_tensor=True)
    sims = [util.cos_sim(embeds[i], embeds[i+1]) for i in range(len(embeds)-1)]
    return float(torch.stack(sims).mean())
```
