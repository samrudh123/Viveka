```python
import scipy
# Example small model (n=4 states, V=2 tokens) that satisfies the constraints.
n = 3
V = 3
e = 0
a = 0.01
b = (1- a)/2
x = 0.01
y = 1 - 2*x

# We construct T^0 and T^1 so that T^0 + T^1 is row-stochastic (rows sum to 1).
T0 = np.array([
    [a*y, b*x, b*x],
    [a*x, b*y, b*x],
    [a*x, b*x, b*y]
])

T1 = np.array([
    [b*y, a*x, b*x],
    [b*x, a*y, b*x],
    [b*x, a*x, b*y]
])

T2 = np.array([
    [b*y, b*x, a*x],
    [b*x, b*y, a*x],
    [b*x, b*x, a*y]
])

w, vl, vr = scipy.linalg.eig((T0+T1+T2), left=True)
idx = np.where(w==1)

model = MarkovMealyModel(n=n, V=V, T_list=[T0, T1, T2])

# By specification the default eta^0 is uniform
print("Initial eta^0 =", vl[1])

tokens, states = model.sample_sequence(max_new_tokens=int(1e4), seed=42)

print("Generated tokens:", tokens)
print("States (eta^t) traversed:")
```

```python
import matplotlib.pyplot as plt
def project_to_simplex(states):
    a = np.array(states)
    x = np.array([-1/np.sqrt(2), 1/np.sqrt(2), 0]).reshape((3, 1))
    y = np.array([-1/np.sqrt(6), -1/np.sqrt(6), 2/np.sqrt(6)]).reshape((3, 1))
    print(a.shape, x.shape, y.shape)
    X = a@x
    Y = a@y
    return np.hstack([X, Y])
out = project_to_simplex(states)
print(out.shape)

plt.scatter(out[:, 0], out[:, 1], c = np.array(states), s=0.1)
plt.show()
```
