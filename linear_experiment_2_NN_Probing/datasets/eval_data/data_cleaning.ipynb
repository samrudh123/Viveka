```python
import json

def remove_duplicate_answers_correctly(data):
    """
    Removes duplicate answers from the 'generated_answers' list for each 
    question and preserves the corresponding 'ground_truth_labels'.

    Args:
        data: The JSON data loaded as a Python dictionary.

    Returns:
        The modified JSON data with duplicate answers and their corresponding
        labels correctly handled.
    """
    for question, details in data.items():
        if "generated_answers" not in details:
            continue

        unique_answers = []
        new_labels = []
        seen_answers = set()
        
        original_answers = details["generated_answers"]
        original_labels = details.get("ground_truth_labels", [])

        # Iterate through answers with their original index
        for i, answer in enumerate(original_answers):
            if answer not in seen_answers:
                seen_answers.add(answer)
                unique_answers.append(answer)
                
                # Use the original index 'i' to append the correct label
                if i < len(original_labels):
                    new_labels.append(original_labels[i])

        details["generated_answers"] = unique_answers
        details["ground_truth_labels"] = new_labels
            
    return data

# Load the JSON data from the file
with open('negated_popqa.json', 'r') as f:
    halueval_data = json.load(f)

# Remove duplicate answers using the corrected function
modified_data = remove_duplicate_answers_correctly(halueval_data)

# Print the modified data to verify
# print(json.dumps(modified_data, indent=2))

with open('negated_popqa.json', 'w') as f:
    json.dump(modified_data, f, indent=2)
```
